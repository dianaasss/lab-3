# Задание 1. Построение простого графика.
Напишите код, который строит график функции y = x^2 на интервале [−10, 10].
Добавьте заголовок, подписи осей и сетку.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(-10, 10, 50)
y = x**2
plt.title("График функции y = x^2")
plt.xlabel("Ось x")
plt.ylabel("Ось y")
plt.grid()
plt.plot(x, y)
plt.show()
```

# Задание 7. Трехмерный график поверхности.
Используя mpl_toolkits.mplot3d, постройте 3D-график функции z = sin(√x2 + y2) на сетке значений x, y в диапазоне [-5, 5].

```python
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='inferno')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

# Задание 8. Множественные подграфики (subplots).
Постройте четыре графика в одной фигуре (2 * 2 сетка):
- Линейный график y = x
- Парабола y = x^2
- Синус y = sin(x)
- Косинус y = cos(x)
Добавьте заголовки в каждому подграфику

```python
x = np.linspace(-10, 20, 100)
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
axs[0, 0].plot(x, x, 'b-')
axs[0, 0].set_title('Линейная функция: y = x')
axs[0, 0].grid(True)
axs[0, 1].plot(x, x**2, 'r-')
axs[0, 1].set_title('Парабола: y = x²')
axs[0, 1].grid(True)
axs[1, 0].plot(x, np.sin(x), 'g-')
axs[1, 0].set_title('Синусоида: y = sin(x)')
axs[1, 0].grid(True)
axs[1, 1].plot(x, np.cos(x), 'm-')
axs[1, 1].set_title('Косинусоида: y = cos(x)')
axs[1, 1].grid(True)
plt.show()
```

# Индивидуальное задание 2

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np
categories = ['Игры', 'Соцсети', 'Образование', 'Музыка', 'Финансы']
downloads = [50, 40, 30, 20, 10]
colors = list(mcolors.BASE_COLORS.values())[:len(categories)]
plt.figure(figsize=(8, 5))
bars = plt.bar(categories, downloads, color=colors, edgecolor='black', alpha=0.7)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height,
             f'{height} млн',
             ha='center', va='bottom')
plt.title('Скачивания приложений', pad=15)
plt.xlabel('Категории')
plt.ylabel('Скачивания (млн)')
plt.grid()
plt.show()
```

# Индивидуальное задание 3

```python
a = 0
b = 4
x = np.linspace(a, b, 200)
y = np.sqrt(x)
under = np.trapz(y, x)
print("Площадь под графиком:", under)
plt.figure(figsize=(8, 5))
plt.plot(x, y, 'b-', linewidth=2)
plt.fill_between(x, y, where=(y>=0), color='skyblue', alpha=0.4)
plt.title('Площадь под кривой $f(x) = \sqrt{x}$ на [0, 4]')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.xlim(a, b)
plt.ylim(0, np.max(y)*1.1)
plt.show()
```

# Индивидуальное задание 4

```python
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.linspace(-2*np.pi, 2*np.pi, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) + np.cos(Y)
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_title("$f(x, y) = \sin(x) + \cos(y)$", fontsize=14)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)
ax.set_xlim(-2*np.pi, 2*np.pi)
ax.set_ylim(-2*np.pi, 2*np.pi)
plt.show()
```
